### Solution

This is just a suggestion, you can improve this by parsing more attributes!

```python
def read_text_from_root(root):
    entries = root.findall(".//zeile")
    morphs = [[]]
    annotations = [[]]

    i = 0
    for entry in entries:
        tokens = entry.findall(".//wortform")
        for token in tokens:
            child_morph = token.find("morph")
            child_annotation= token.find("annotation")
            annotations[i].append(child_annotation.get("gefunden"))
            morphs[i].append(child_morph.text)
        i += 1
        annotations.append([])
        morphs.append([])
    return annotations
```
